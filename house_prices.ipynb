# Setup data.gov.uk Sparql query

```python
query = """
prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#>
prefix owl: <http://www.w3.org/2002/07/owl#>
prefix xsd: <http://www.w3.org/2001/XMLSchema#>
prefix sr: <http://data.ordnancesurvey.co.uk/ontology/spatialrelations/>
prefix lrhpi: <http://landregistry.data.gov.uk/def/hpi/>
prefix lrppi: <http://landregistry.data.gov.uk/def/ppi/>
prefix skos: <http://www.w3.org/2004/02/skos/core#>
prefix lrcommon: <http://landregistry.data.gov.uk/def/common/>

# Returns the Price Paid data from the default graph for each transaction record having
# an address with the given postcode.
# The postcode to query is set using SPARQL 1.1's 'values' clause

SELECT ?paon ?saon ?street ?town ?county ?postcode ?amount ?date ?category
WHERE
{{
  VALUES ?postcode {{"{postcode_var}"^^xsd:string}}

  ?addr lrcommon:postcode ?postcode.

  ?transx lrppi:propertyAddress ?addr ;
          lrppi:pricePaid ?amount ;
          lrppi:transactionDate ?date ;
          lrppi:transactionCategory/skos:prefLabel ?category.

  OPTIONAL {{?addr lrcommon:county ?county}}
  OPTIONAL {{?addr lrcommon:paon ?paon}}
  OPTIONAL {{?addr lrcommon:saon ?saon}}
  OPTIONAL {{?addr lrcommon:street ?street}}
  OPTIONAL {{?addr lrcommon:town ?town}}
}}
ORDER BY ?amount
"""
```

# Run Query and import into Pandas Dataframe

```python
from SPARQLWrapper import SPARQLWrapper, JSON

sparql = SPARQLWrapper("http://landregistry.data.gov.uk/landregistry/query")

non_empty_postcode = 'PL6 8RU'
postcodes = [non_empty_postcode] + postcodes
for p_idx, postcode in enumerate(postcodes):
    query_str = query.format(postcode_var=postcode)
#     print(query_str)
    sparql.setQuery(query_str)
    
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()
#     print(results)
    if p_idx == 0:
        keys = results["results"]["bindings"][0].keys()
        df = pd.DataFrame(columns=(keys))
        continue
        
    for i, row in enumerate(results["results"]["bindings"], start=df.size):
        df.loc[i] = [row[key]['value'] for key in keys]
# print(results["results"]["bindings"][0].keys)
# make sure there is nothing from plymouth in here
df=df[df.postcode.str.contains("PL6") == False]

# for street in ["ST EDYTHS ROAD",]:
#     df=df[df.street.str.contains(street) == False]
df=df[df.street.str.contains("ST EDYTHS ROAD") == False]
df=df[df.paon.str.contains("SEA MILLS HOUSE") == False]
# remove 3 BOWERLEAZE 
df = df[~((df.street.str.contains("BOWERLEAZE")) & (df.paon.str.contains('3')))]
```

# Linear Regression

```python
import pandas as pd
from statsmodels.formula.api import ols
import statsmodels.api as sm
import numpy as np
```

## Change date time index to an int  value: days since start of records

```python
x = df.index.to_series()
df['date_delta'] = x-x.iloc[0]
```

```python
df['delta']=(df['date_delta'] / np.timedelta64(1, 'D')).astype(int)
```

```python
df.tail()
```

## Manual OLS

```python
import statsmodels.api as sm
X = df['delta']
y=  df['amount']
X = sm.add_constant(X)

m_model = sm.OLS(y,X).fit()

b = m_model.params[0]
m = m_model.params[1]

m_model.summary()
```

```python
last_row = df.ix[-1]
amount = last_row['delta'] * m + b
amount
```

## Formula based OLS

```python
f_model = ols(formula="amount ~ delta", data=df).fit()
f_model.summary()
```

```python
m = f_model.params[1]
b = f_model.params[0]
amount = last_row['delta'] * m + b
amount
```
